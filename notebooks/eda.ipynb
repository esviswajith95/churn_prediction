# Telecom Churn Prediction - Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
data.info()
```

```python
data.head()
```

## Categrical variables

### Distributions

```python
features = data.columns
fig = plt.figure(figsize=(20, 16))
pos = 1
for feature in features:
    if feature not in ['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']:
        ax = plt.subplot(5, 4, pos)
        sns.countplot(data=data, x=feature, palette='viridis',  ax=ax)
        plt.xlabel('Values')
        plt.title(feature)
        pos += 1
plt.subplots_adjust(wspace=0.5, hspace=0.5)

plt.suptitle("Distribution of categorical features")
plt.show()
```

## Numerical variables

### Tenure

```python
sns.histplot(data=data, x = 'tenure')
```

### Monthly charges

```python
sns.histplot(data=data, x = 'MonthlyCharges')
```

### Total Charges

```python
data['TotalCharges'] = data['TotalCharges'].replace(' ', '0')
data['TotalCharges'] = data['TotalCharges'].astype('float')
sns.histplot(data=data, x = 'TotalCharges')
```
